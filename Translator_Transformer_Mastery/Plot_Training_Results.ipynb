# **Plot Training Results**

```python
from pickle import load
from matplotlib.pylab import plt
from numpy import arange

train_loss = load(open('./data/train_loss.pkl', 'rb'))
val_loss = load(open('./data/val_loss.pkl', 'rb'))

train_values = train_loss.values()
val_values = val_loss.values()


train_values = [e.numpy() for e in train_values]
val_values = [e.numpy() for e in val_values]


epochs = range(len(train_values))


plt.plot(epochs, train_values, label='Training Loss')
plt.plot(epochs, val_values, label='Validation Loss')


plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(arange(0, len(train_values), 2))

plt.legend(loc='best')
plt.show()
```
